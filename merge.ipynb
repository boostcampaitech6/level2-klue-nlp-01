```python
from transformers import AutoTokenizer, AutoConfig, AutoModelForMaskedLM, AutoModelForSequenceClassification, AutoModel
import torch
MODELNAME = 'klue/roberta-large'
tokenizer = AutoTokenizer.from_pretrained(MODELNAME)
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
# MLM
model_mlm = AutoModelForMaskedLM.from_pretrained(MODELNAME, max_position_embeddings=514)#, vocab_size=40723)
model_mlm
```

```python
# classification
model_classification = AutoModelForSequenceClassification.from_pretrained(MODELNAME, num_labels=30, max_position_embeddings=514)
model_classification
```

```python
model_mlm_pre = AutoModelForMaskedLM.from_pretrained(MODELNAME, max_position_embeddings=514)#, vocab_size=40723)
model_mlm_pre.load_state_dict(torch.load('/data/ephemeral/roberta-large-pretrained/roberta-large-64-5e-05_pre.pt'))
model_mlm_pre
```

```python
import copy
model_final = copy.deepcopy(model_classification)
pretrained_dict = model_mlm_pre.state_dict()  # pretrained 상태 로드
model_dict = model_classification.state_dict() # 현재 신경망 상태 로드
# 1. filter out unnecessary keys
pretrained_dict = {k: v for k, v in pretrained_dict.items() if k in model_dict}
# 2. overwrite entries in the existing state dict
model_dict.update(pretrained_dict) 
# 3. load the new state dict
model_final.load_state_dict(model_dict)
```

```python
# 1과 2가 다르고, 2와 4가 같고, 5가 달라야 함
print("MLM from transformers         : ", model_mlm.state_dict()['roberta.encoder.layer.1.output.LayerNorm.bias'][:5])
print("Pretrained MLM                : ", model_mlm_pre.state_dict()['roberta.encoder.layer.1.output.LayerNorm.bias'][:5])
print("Classifier from transformers  : ", model_classification.state_dict()['roberta.encoder.layer.1.output.LayerNorm.bias'][:5])
print("Adaptive pretrained Classifier: ", model_final.state_dict()['roberta.encoder.layer.1.output.LayerNorm.bias'][:5])
model_final_fine = torch.load('/data/ephemeral/parameters/roberta-large-64-5e-05_tapt.pt')
print("Final Classifier              : ", model_final_fine['roberta.encoder.layer.1.output.LayerNorm.bias'][:5])
```
